9) Which line of Python code would correctly calculate the average vote percentage (%) for candidates who lost in the election?
- a. elections.groupby('Result')['%'].median()
- b. elections.loc[0]['%'].mean()
- c. elections.groupby('Result') = 'loss' ['%'].mean()
- d. elections[elections['Result'] == 'loss']['%'].mean() (correct)
- e. elections.groupby('Result') == 'loss'.mean()

original Question  j) What is the average vote percentage for candidates who won their election? Is this average higher compared with candidates who lost? <br> k) Which candidates received the highest and lowest vote percentages? Display the candidate names along with their vote percentages.

```python
# Average vote % for winners
average_winner = elections[elections['Result'] == 'win']['%'].mean()

# Average vote % for losers
average_loser = elections[elections['Result'] == 'loss']['%'].mean()

print(f"Average vote % for winners: {average_winner:.2f}")
print(f"Average vote % for losers: {average_loser:.2f}")

```

<b>Measurement Question 10 <br>

10) Which candidate has the lowest vote percentage? Display the candidate's name and their % vote count
- a. Romney with 29.14%
- b. Dole with 15.07%
- c. McCain with 7.51% 
- d. Kerry with 10.46% 
- e. Anderson with 6.6% (correct)



```python
# Candidate with the highest vote %
highest_vote = elections['%'].idxmax()
highest_vote_value = elections.loc[highest_vote, '%']

# Candidate with the lowest vote %
lowest_vote = elections['%'].idxmin()
lowest_vote_value = elections.loc[lowest_vote, '%']

print(f"Highest vote: {highest_vote} with {highest_vote_value}%")
print(f"Lowest vote: {lowest_vote} with {lowest_vote_value}%")
```
